## **Ejercicio 4**

Un cajero autom√°tico tiene 500.000 pesos disponibles.
Un usuario puede retirar varias veces, pero no puede retirar m√°s de lo que queda en el cajero. Escribe un programa que:

- Pregunte cu√°nto quiere retirar.

- Reste ese valor al saldo.

- Termine cuando el saldo llegue a 0 o el usuario escriba 0.

```python
### TU C√ìDIGO AQU√ç ###
saldo = 500000

while saldo > 0:
    retiro = int(input(f"Saldo disponible: {saldo}. ¬øCu√°nto desea retirar? (0 para salir): "))
    
    if retiro == 0:
        break
    
    if retiro > saldo:
        print("No puede retirar m√°s de lo disponible.")
    else:
        saldo -= retiro
        print(f"Retiro exitoso. Nuevo saldo: {saldo}")

print("Operaci√≥n finalizada.")
```

## **Ejercicio 5**

Crea una funci√≥n raiz_cuadrada(num) que calcule la ra√≠z cuadrada de un n√∫mero.

Si el n√∫mero es negativo, devuelve: "Error: no se puede calcular la ra√≠z cuadrada de un n√∫mero negativo."

Si el valor no es num√©rico, devuelve: "Error: el valor debe ser num√©rico."

NOTA: utiliza import math y la funci√≥n math.sqrt(num). No olvides documentar la funci√≥n.

```python
### TU C√ìDIGO AQU√ç ###
import math

def raiz_cuadrada(num):
    """
    Calcula la ra√≠z cuadrada de un n√∫mero.
    Devuelve error si el n√∫mero es negativo o no es num√©rico.
    """
    
    if not isinstance(num, (int, float)):
        return "Error: el valor debe ser num√©rico."
    
    if num < 0:
        return "Error: no se puede calcular la ra√≠z cuadrada de un n√∫mero negativo."
    
    return math.sqrt(num)

```

```python
print(raiz_cuadrada(16))
```

```python
print(raiz_cuadrada(-9))
```

## **Ejercicio 6**

Escribe una funci√≥n llamada es_par que reciba un n√∫mero entero como par√°metro y retorne True si el n√∫mero es par y False si es impar.

NOTA: % es el operador m√≥dulo en Python. Lo que hace es dividir un n√∫mero entre otro y devolver el residuo (lo que sobra de la divisi√≥n entera).

Si al dividir un n√∫mero en 2, el residuo es cero, entonces el n√∫mero es par, de lo contrario, es impar.

```python
### TU C√ìDIGO AQU√ç ###
def es_par(numero):
    """
    Determina si un n√∫mero entero es par.

    Par√°metros:
    numero (int): N√∫mero entero a evaluar.

    Retorna:
    bool: True si el n√∫mero es par, False si es impar.
    """
    if numero % 2 == 0:
        return True
    else:
        return False
```

```python
print(es_par(10))   # Deber√≠a imprimir True
```

```python
print(es_par(7))    # Deber√≠a imprimir False
```

## **Ejercicio 7**

Escribe un c√≥digo que permita al usuario crear una contrase√±a y que verifique que esta cumpla todas las siguientes condiciones:

- Tiene al menos 8 caracteres.

- Contiene al menos una letra may√∫scula.

- Contiene al menos una letra min√∫scula.

- Contiene al menos un n√∫mero.

Si cumple todas las condiciones, muestra "Contrase√±a v√°lida", de lo contrario, explica qu√© condici√≥n no se cumple.

```python
### TU C√ìDIGO AQU√ç ###
def verificar_contrasena(contrasena):
    if len(contrasena) < 8:
        print("Debe tener al menos 8 caracteres.")
        return
    
    if not any(c.isupper() for c in contrasena):
        print("Debe contener al menos una letra may√∫scula.")
        return
    
    if not any(c.islower() for c in contrasena):
        print("Debe contener al menos una letra min√∫scula.")
        return
    
    if not any(c.isdigit() for c in contrasena):
        print("Debe contener al menos un n√∫mero.")
        return
    
    print("Contrase√±a v√°lida.")

verificar_contrasena("Hola1234")
```

usuario@dominio.com
    
Por ejemplo: si se le pasa "lmcastroco@gmail.com", la funci√≥n retornar√°: gmail.com

```python
### TU C√ìDIGO AQU√ç ###
def extraer_dominio(email):
    """
    Extrae el dominio de un correo electr√≥nico.

    Par√°metros:
    email (str): Direcci√≥n de correo en formato usuario@dominio.

    Retorna:
    str: Dominio del correo electr√≥nico.
    """
    partes = email.split("@")
    return partes[1]
```

## **Ejercicio 10**

Vas conduciendo un carro muy r√°pido y eres detenido por un agente de tr√°nsito.

Escribe una funci√≥n "multa" que retorne uno de los 3 siguientes resultados:  

- Si la velocidad es 60 Km/h o menos, el resultado es "Sin multa".
- Si la velocidad est√° entre 60 y 80 Km/h (inclusive), el resultado es "Multa peque√±a".
- Si la velocidad es mayor a 80 Km/h, el resultado es "Multa grande".

Adem√°s, considera que si esto ocurre el d√≠a de tu cumplea√±os (codificado como par√°metro booleano que recibe la funci√≥n), tu velocidad podr√≠a ser 5 Km/ mayor en todos los casos.

```python
### TU C√ìDIGO AQU√ç ###
def multa(velocidad, es_cumpleanos):
      
      """
    Determina el tipo de multa seg√∫n la velocidad del veh√≠culo.

    Par√°metros:
    velocidad (int): Velocidad a la que iba el conductor.
    es_cumpleanos (bool): True si es cumplea√±os del conductor.

    Retorna:
    str: Tipo de multa ('Sin multa', 'Multa peque√±a', 'Multa grande').
    """

    if es_cumpleanos:
        velocidad -= 5
    
    if velocidad <= 60:
        return "Sin multa"
    elif velocidad <= 80:
        return "Multa peque√±a"
    else:
        return "Multa grande"
```

```python
multa(81, True)
```

```python
multa(81, False)
```

## **Ejercicio 14**

Tienes un diccionario con exportaciones en millones de USD. Escribe un ciclo for que recorra los valores, calcule el promedio de exportaciones y lo imprima como: "El promedio de exportaciones es: XXXX millones de USD"

```python
exportaciones = {"Colombia": 1500, "M√©xico": 2300, "Chile": 1200, "Per√∫": 1800}
```

## **Ejercicio 16**

Una empresa quiere vender al menos 100 unidades de un producto. Cada d√≠a vende un n√∫mero aleatorio de unidades entre 1 y 10 (pista: usa import random y random.randint(1,10)). Escribe un programa con while que muestre:

- El total vendido cada d√≠a.

- En cu√°ntos d√≠as se logr√≥ la meta.

```python
### TU C√ìDIGO AQU√ç ###
import random

total_vendido = 0
dias = 0

while total_vendido < 100:
    ventas_dia = random.randint(1, 10)
    total_vendido += ventas_dia
    dias += 1

print(f"Meta alcanzada en {dias} d√≠as con un total de {total_vendido} unidades vendidas.")
```
